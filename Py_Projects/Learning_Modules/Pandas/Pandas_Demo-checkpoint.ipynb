```python
df.rename(columns={'ConvertedComp': 'SalaryUSD'}, inplace=True)
df
```

```python
df['Hobbyist'] = df['Hobbyist'].map({'Yes': True, 'No': False})
```

```python
# Here we create a filter to see which developement languages have the highest earning potential.
# we simply save the columns as a list in the second arg space of loc
high_salary_filter = (df['ConvertedComp']) > 70000
df.loc[high_salary_filter, ['Country','LanguageWorkedWith', 'ConvertedComp']]
```

```python
# Here we create a similiar filter to the one above, however in this filter we 
# use want to specify which specific countries within the Country column. So we
# set a variable equal to a list of the country names and setup a filter
# variable equal to the Country column values if they are in country_list
# finally this is all passed to loc as the row arg, and 'Country' as the column arg

country_list = ['United States', 'Canada', 'India', 'United Kingdom']
filt = df['Country'].isin(country_list)
df.loc[filt, 'Country']
```

```python
# We setup a filter here to check the strings in the column 'LanguageWorkedWith'
# to see if the string value 'Python' in included in the string. Also, we can pass
# na=False as a second argument which accounts for exceptions
# of NAN values in the dataset
filt = df['LanguageWorkedWith'].str.contains('Python', na=False)
df.loc[filt, 'LanguageWorkedWith']
```

```python
no_py_users = str(len(df.loc[filt, 'LanguageWorkedWith']))
print('No. of People who used Python:'+ ' ' + no_py_users)
```

```python
# Here we run the value_counts() method on the column hobbyist to
# first return the number of yes and no responses holistically,
# then we pass normalize=True keywarg to the value_counts() method 
# to get the number of yes and no expressed as a percentage
# df['Hobbyist'].value_counts()
df['Hobbyist'].value_counts(normalize=True)
```

```python
df['Country'].value_counts()
```

```python
# Here we use the get_group() method to return the 
# dataframe in groups seperated by 'Country' name.
# The key difference between this and a filter like
# we built befor is that now we are able to access all
# of  the data which is split into groups instead of only the group where the condition is met
country_group = df.groupby(['Country'])
country_group.get_group('United States')
```

```python
# This is how we were filtering the countries, the key note is 
# that once we do a conditional filter we are changing the dataframe into a
# new object without the other country elements of the original dataframe.
filt = df['Country'] == 'United States'
df.loc[filt]['Hobbyist'].value_counts()
```

```python
# Here we run head on the grouped column to see value counts
# for the first 25 countries
country_group['Hobbyist'].value_counts().head(25)
```

```python
# Here we use loc to find the value_counts() for the United States
country_group['Hobbyist'].value_counts().loc['United States']
```

```python
# Here we get the normalized value_counts for the Russian Federation
country_group['Hobbyist'].value_counts(normalize=True).loc['Russian Federation']
```

```python
# Here we run the median method on the grouped converted comp column
# for the Canada rows
country_group['ConvertedComp'].median().loc['Canada']
```

```python
# Here we run the agg method to call 2 seperate functions on the 
# groupedby converted comp canada rows
country_group['ConvertedComp'].agg(['median', 'mean']).loc['Canada']
```

```python
# Here we use a filter to check if the country name == 'India'
# then we pass the filter to loc along with the column of languages worked with for 
# India, then we run the str method contains on the values of language worked with to see
# if they contain 'Python'. Finally we return the sum value to count the number of Trues produced by
# contains
filt = df['Country'] == 'India'
df.loc[filt]['LanguageWorkedWith'].str.contains('Python').sum()
```

```python
# Here python throws an error when trying to run this line.
# The issue is that we cannot use the str method on the series groupby object
# country_group. See next cell for solution.
country_group['LanguageWorkedWith'].str.contains('Python')
```

```python
# Here we use the apply method to apply a lambda function which calls the string method
# .contains() then sums the series to see where True for Python in string.
# sum counts True as a 1 and false as a 0 and sums all instance for final count.

country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
# Challenge: Produce a dataframe with the percentage of people from each country who know python.
#################################################################################################
```

```python
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
# Create a dataframe variable of the language worked with  by using the approach above. 
country_uses_python = country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python
# By default we concat will concatenate these dataframes on rows,but
# we want to match based on columns so we set axis='columns'
python_df = pd.concat([country_respondents, country_uses_python], axis='columns', sort=False)
python_df.rename(columns={'Country': 'Total No Coders', 'LanguageWorkedWith': 'No of Python Coders'}, inplace=True)
python_df
```

```python
python_df['% KnowsPython'] = (python_df['No of Python Coders'] / python_df['Total No Coders']) * 100
python_df
```

```python
python_df.sort_values(by='% KnowsPython', ascending=False, inplace=True)
python_df.head(50)
```
